**Fermi-Dirac Distribution**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate energies
E = np.linspace(0, 10, 2000)

# Fermi–Dirac distribution
def fermi_dirac(E, T, Ef=5):
    kB = 8.617e-5  # eV/K
    return 1 / (np.exp((E - Ef) / (kB * T)) + 1)

# Calculate distributions
f0 = fermi_dirac(E, 0.1)   # ~0 K
f293 = fermi_dirac(E, 293)
f2000 = fermi_dirac(E, 2000)

# Plot
plt.figure(figsize=(8, 4))

# Linewidth
lw = 1.5

plt.plot(E, f0, label="0 K", color="black", linewidth=lw)
plt.plot(E, f293, label="293 K", color="red", linewidth=lw)
plt.plot(E, f2000, label="2000 K", color="dodgerblue", linewidth=lw)

plt.xlabel("Electron energy [eV]")
plt.ylabel("Probability of occupation")

# plt.title("Fermi–Dirac distribution for several temperatures", pad=20)

plt.xlim(0, 10)
plt.ylim(0, 1.2)

plt.legend()
plt.grid(False)

# # ---- Add custom xtick at E_F = 5 eV ----
# ticks = list(plt.xticks()[0])          # get current tick locations
# if 5 not in ticks:
#     ticks.append(5)
# ticks = sorted(ticks)

# labels = [f"{t:g}" for t in ticks]     # default numeric labels
# for i, t in enumerate(ticks):
#     if t == 5:
#         labels[i] = r"$E_F$"           # label at 5 eV

# plt.xticks(ticks, labels)
# # ----------------------------------------

# ---- Replace all xticks with only E_F ----
plt.xticks([5], [r"$E_F$"])
# ------------------------------------------

# Save figure
plt.savefig("fermi_dirac.png", dpi=300, bbox_inches="tight")
plt.show()
```
